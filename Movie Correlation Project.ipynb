```python
#import libeaeies

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjust the configuration we will create

#Read the data

df = pd.read_csv(r'C:\Users\Admin\Downloads\movies\movies.csv') 





```

```python
#lets look at the data

df.head()
```

```python
#lets try if there any missing datas


for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(('{} - {}%'.format(col,pct_missing))
```

```python
df.dtypes
```

```python
#changing data types of columns

df['budget']=df['budget'].astype('int64')
```

```python
df['budget']
```

```python
df.drop_duplicates()
```

```python
#budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.show()
```

```python
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross')
```

```python
df.corr(method='pearson')
```

```python
correlation_matrix= df.corr(method='kendall')
sns.heatmap(correlation_matrix,annot=True)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
df_numerised=df
for col_name in df_numerised.columns:
    if (df_numerised[col_name].dtype=='object'):
        df_numerised[col_name]=df_numerised[col_name].astype('category')
        df_numerised[col_name]=df_numerised[col_name].cat.codes
df_numerised
        
```

```python
correlation_matrix= df_numerised.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
correlation_mat=df_numerised.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
```

```python
sorted_pairs
```

```python
high_corr=sorted_pairs[sorted_pairs>0.5]
high_corr
```
